```python
import folium

map1 = folium.Map(location=[17.3850,78.4867], zoom_start=12)
folium.Marker(location=[17.3850,78.4867],
              popup='SR University, Telangana',tooltip='Click for info').add_to(map1)

# Example definition for locations
locations = [
    [17.3850, 78.4867, 'SR University, Telangana'],
    [17.4000, 78.5000, 'Another Point A'],
    [17.3700, 78.4700, 'Another Point B']
]

for loc in locations:
    folium.Marker(location=[loc[0], loc[1]], popup=loc[2], icon=folium.Icon(color='green', icon='ok-sign')).add_to(map1)

map1
```

```python
import folium
import pandas as pd
data=pd.DataFrame({
    'State':['Telangana','andhra pradesh','Karnataka'],
    'Cases': [2300, 3100, 2800]
})
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/world-countries.json'

map3 = folium.Map(location=[15.9129, 79.7400], zoom_start=5)

folium.Choropleth(
geo_data=url,
data=data,
columns=['State', 'Cases'],
key_on='feature.properties.name',
fill_color='YlOrRd',
legend_name='COVID-19 Cases'
).add_to(map3)
```

```python
import plotly.express as px
data = px.data.gapminder().query("year == 2007")
fig = px.scatter_geo(
data,
locations="iso_alpha",
color="continent",
hover_name="country",
size="pop",
projection="natural earth",
title="World Population by Country (2007)"
)
fig.show()
```

```python
import plotly.express as px
import pandas as pd
data = pd.DataFrame({
'Country': ['India', 'USA', 'UK', 'Brazil', 'Japan'],
'iso_alpha': ['IND', 'USA', 'GBR', 'BRA', 'JPN'],
'Vaccination_Rate': [78, 65, 80, 60, 75]
})
fig = px.choropleth(
data,
locations="iso_alpha",
color="Vaccination_Rate",
hover_name="Country",
color_continuous_scale=px.colors.sequential.Plasma,
title="COVID-19 Vaccination Rate by Country"
)
fig.show()
```

```python
import plotly.express as px
# Sample dataset
df = px.data.carshare()
fig = px.density_mapbox(
df,
lat='centroid_lat',
lon='centroid_lon',
z='car_hours',
radius=10,
center=dict(lat=40.7, lon=-74),
zoom=10,
mapbox_style="stamen-terrain",
title="Car Sharing Density Map"
)
fig.show()
```
